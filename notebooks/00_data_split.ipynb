```python
df['date'] = pd.to_datetime(df['date'])
df['date'].dt.year.value_counts().sort_index()
```

<h3> It's noted that we have the data for <b>10 years</b> and this is the amount of each data per year so overall the dataset seems to be balanced</h2>

Avoid Leakage
- Train set (2012-2019)
- Validation/Eval set (2020-2021)
- Test set (2022-2023)

```python
df = df.sort_values(by='date')
```

```python
df.head()
```

```python
#Define Cutoff dates
cutoff_date_eval = '2020-01-01'
cutoff_date_test = '2022-01-01'
```

```python
# Train: before 2020
train_df = df[df["date"] < cutoff_date_eval]

# Validation/Eval: 2020–2021
eval_df = df[(df["date"] >= cutoff_date_eval) & (df["date"] < cutoff_date_test)]

# Holdout: 2022–2023
test_df = df[df["date"] >= cutoff_date_test]

print("Train shape:", train_df.shape)
print("Eval shape:", eval_df.shape)
print("test shape:", test_df.shape)
```

# 3. Extracting Data

```python
train_df.to_csv('../data/raw/train.csv', index=False)
eval_df.to_csv('../data/raw/eval.csv', index=False)
test_df.to_csv('../data/raw/test.csv', index=False)
```
